```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Read data from CSV file
data = pd.read_csv('data.csv')

def draw_raw():
    # Create box plots
    fig, ax= plt.subplots(1, 1, figsize=(4, 3))
    ax.boxplot([data['SIL_N'], data['SIL_B'], data['SIL'], data['AMI']])
    ax.set_xticks([1, 2, 3, 4], ['SIL_N', 'SIL_B', 'SIL', 'AMI'])
    ax.set_xlabel('Metrics')
    ax.set_ylim([0, 1])
    ax.set_ylabel('Values')
    fig.tight_layout()
    return fig, ax

fig, ax = draw_raw()
fig.savefig('metrics.svg')
```

```python
fig, ax = draw_raw()


# Create a rectangle patch
rect = Rectangle((0.6, 0.05), 3, 0.9, alpha=0.2, facecolor='red')
# Add the rectangle patch to the figure
ax.add_patch(rect)
# Display the updated figure

fig.savefig('metrics_1.svg')
```

```python
fig, ax = draw_raw()


# Create a rectangle patch
rect = Rectangle((3.6, 0.3), 0.8, 0.9, alpha=0.2, facecolor='blue')
# Add the rectangle patch to the figure
ax.add_patch(rect)
# Display the updated figure

fig.savefig('metrics_2.svg')
```
